```python
import pandas as pd
```

```python
df = pd.DataFrame.from_csv('../data/curves.csv', index_col = None)
df_ref = pd.DataFrame.from_csv('../data/reference.csv', index_col = None)
```

```python
df.head(n=5)
```

```python
df_ref.head(n=5)
```

```python
df_cln = df.copy()
orgs = list(df['Orgs'].values)
new_orgs = []
k = [str(x) for x in df_ref['ID'].values]
id_dict = dict(zip(k, df_ref['4-6 LETTER CODE'].values))
for org in orgs:
    org = org.split('_')
    
    if org[0] in id_dict.keys():
        org[0] = id_dict[org[0]]
    if org[0] in id_dict.values():
        new_orgs.append('_'.join(org))
    else:
        new_orgs.append('NULL')
    
df_cln['Orgs'] = new_orgs
df_cln = df_cln[df_cln['Orgs']!='NULL']
df_cln.to_csv('curves_cln.csv', index = False)
```
